```r
df %>% 
        mutate(temp = case_when(
            
            Temperature < 20 ~ "Low",
            Temperature >=20 & Temperature <=22 ~ "Mild",
            TRUE ~ "High"
        )) -> p_df

    
    
    plot_bar(p_df,
             by ='temp',
              ggtheme = theme_bw())
```

```r
df %>% 
    mutate(Occupancy = as.factor(Occupancy)) %>% 
    select_if(is.numeric) -> df_numeric
    
plot_correlation(na.omit(df_numeric))
```
